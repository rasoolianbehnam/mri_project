# Compute aggregate metrics for all muscles from pre-computed datasets

```python
%reset -f
%load_ext autoreload
%autoreload 2
#matplotlib figure size
from pylab import rcParams
rcParams['figure.figsize'] = 10, 10
from custom_imports import *
```

```python
df = pd.read_csv("../data/prediction_stats_v01.csv")

res = df.groupby(['metric', 'muscle'])[['value']].mean()# .reset_index().set_index('metric')
res = res.unstack(1)['value']
res.loc['90_degree_lever_arm'] = res.loc['lever_arm_89.8_distance']
res.drop('lever_arm_89.8_distance', axis=0)
res.index = res.index.map(str.title).map(lambda x: x.replace('_', ' '))
res = res.sort_values(-1, na_position='first').round(2).fillna('N/A')
res['all'] = res[-1]
res = res.drop(-1, axis=1)
res
```

```python
df = pd.read_csv("../data/prediction_stats_v02.csv").fillna(0)

res = df.groupby(['metric', 'muscle'])[['value']].mean()# .reset_index().set_index('metric')
res = res.unstack(1)['value']
res.loc['90_degree_lever_arm'] = res.loc['lever_arm_89.8_distance']
res.drop('lever_arm_89.8_distance', axis=0)
res.index = res.index.map(str.title).map(lambda x: x.replace('_', ' '))
res = res.sort_values(-1, na_position='first').round(2).fillna('N/A')
res['all'] = res[-1]
res = res.drop(-1, axis=1)
res = res.drop([0, 12, 14, 16, 18, 19, 22], axis=1).drop('Lever Arm 89.8 Distance')
```

```python
tmp_ = res.loc[['Accuracy', 'Precision', 'Recall']].copy()
tmp_.loc['Accuracy', 'all'] = res.loc['Total Accuracy', 'all']
tmp_.rename(columns={'all': 'Total'})
```

```python
tmp_ = res.loc[['Center Distance', '90 Degree Lever Arm', 'Area Ratio']].copy()
tmp_.drop('all', 1)# rename(columns={'all': 'Total'})
```
